```python
# Section05-1
# BeautifulSoup
# BeautifulSoup 사용 스크랩핑 (1) - 기본 사용법 

from bs4 import BeautifulSoup
 
```

```python
html = """
<html>
  <head>
    <title> The dormouse's story </title>
  </head>
  <body>
    <h1>this is  h1 area</h1>
    <h2>this is h2 area</h2>
    <p class ="title"><b> The Dormos Stroy</b></p>

    <p class ="story">Once upon a thime there  wer three 
      <a href="http://example.com/else" class="sister" id="link1">Elsie</a> 
      <a href="http://example.com/Ee" class="sister" id="link2">Else</a>
      <a data-test="test" data-io="link3" href="http://example.com/littele" class="brother" id="link3">Title</a>
    </p>

    <p class ="story">
      story ...vhjkmvvvbbvnvbnvbn
    </p>

  </body>

</html>

"""

```

```python
# 예제 1 (BeautifulSoup - basic )
# bs4 초기화 
soup = BeautifulSoup(html, 'html.parser')

# 타입 확인 
print('soup',type(soup))
print('prettify',soup.prettify())
print()
print()

# h1 태그 접근 
h1 = soup.html.body.h1
print('h1  = ' ,h1)
print()
print()

# p 태그 접근 
p1 = soup.html.body.p
print('p1',p1)
print()
print()

# 두번째 p 태그 접근 - 한번 호출시 공백으로 간다 
p2 = p1.next_sibling.next_sibling
print('p2 = ',p2)
print()
print()



```

```python
# 텍스트만 출력1 
print('h1 = ' ,h1.string)

# 텍스트만 출력2 - b태그는 제외됨 
print('h1 = ' ,p1.string)

#사용가능한 함수 확인 
print(dir(p2))

# 다음 엘리먼트 확인 
print(list(p2.next_element))
print(p2.next_element)

# 반복 출력 확인 
for v in p2.next_element:
  print(v)


# Beautiful soup  공식 = https://www.crummy.com/software/BeautifulSoup/bs4/doc/

```

```python
# 예제 2 (Find, Find_all)

soup2 = BeautifulSoup(html,'html.parser')

# a 태그 모두 선택 
link1 = soup.find_all('a')
print('links',link1)


link11 = soup.find_all('a',limit=2)
print('links',link11)
print()
print()

# a 태그 중 brother class를 가져오기  
link2 = soup.find_all('a', class_="sister")
print(link2)

link22 = soup.find_all('a', string=["Else"])
print(link22)

link222 = soup.find_all('a', string=["Else","Title"])
print(link222)


link2222 = soup.find_all('a', id="link3")
print(link2222)
print()
print()

for t in link2:
  print(t)

print()
print()

# 처음 발견한 a 태그 선택 - find 는 가장 먼저 발견한 a 태그를 가져온다 
link3 = soup.find("a")

print(link3)
print(link3.string)
print(link3.text)
print()
print()

# 다중조건
link4 = soup.find("a", {"class":"brother", "data-io":"link3"})

print(link4)
print(link4.string)
print(link4.text)
print()
print()
```

```python
# css 선택자 : select , select_one
# 태그로 접근 : find , find_all
# 예제3 (select , select_one) 

# 태그 + 클래스 + 자식선택자 
link5 = soup.select_one('p.title > b') # .은 clas 이름이다 
print(link5)
print(link5.string)
print(link5.text)
print()
print()


link6 = soup.select_one('a#link1') # #은 id 이다 
print(link6)
print(link6.string)
print(link6.text)
print()
print()


link7 = soup.select_one("a[data-io='link3']")
print(link7)
print(link7.string)
print(link7.text)
print()
print()


link77 = soup.select_one("a[data-test='test']")
print(link77)
print(link77.string)
print(link77.text)
print()
print()


# 선택자에 맞는 전체 선택 
link8 = soup.select('p.story > a') # p태그 story 하위 중 a 태그를 모두가져와라
print(link8)
#print(link8.string)

print()
print()


link9 = soup.select('p.story > a:nth-of-type(2)') # story 하위 중 a태그에서 2번째를 가져와라 
# css 선택자 nth-of-type 는 구글참조   https://aboooks.tistory.com/317 
print(link9)
print()
print()

link10 = soup.select("p.story") # 리스트 두개를 가져옴
print(link10)
print()
print()


for t in link10:
  temp = t.find_all("a")

  if temp: # 하위에 a 태그 데이터가 있는 경우 
    for v in temp:
      print(">>>>>",v.string)
  else: # 데이터가 없다면 
      print('x = ' , t)
```
