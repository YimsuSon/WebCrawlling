```python
from selenium import webdriver
import time 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.chrome.options import Options 
from bs4 import BeautifulSoup
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless") # 브라우저 실행 x 모드 설정 

# webdriver 설정 (Chrome, Firefox 등 ) - Headless 모드 브라우저 실행x
browser = webdriver.Chrome('./chromedriver.exe',options=chrome_options)

# webdriver 설정 (Chrome, Firefox 등 ) - 브라우저 실행 모드 
#browser = webdriver.Chrome('./chromedriver.exe')
```

```python
# 크롬 브라우저 내부 대기 
browser.implicitly_wait(3)

# 브라우저 사이즈 
#browser.set_window_size(1920,1280)
```

```python
# 페이지 이동 
browser.get('http://prod.danawa.com/list/?cate=112758&15main_11_02')
```

```python
# 1차 페이지 내용확인 
print('Before Page Contents : {}'.format(browser.page_source))
```

```python
# 제조사별 더 보기 버튼 클릭1 
# Emplicitly wait - direct 접근 

# XPATH  버튼이 나타날때까지 브러우저가 3초까지 기다림. 안나오면 에러발생 
WebDriverWait(browser,3).until(EC.presence_of_element_located((By.XPATH,'//*[@id="dlMaker_simple"]/dd/div[2]/button[1]'))).click()
```

```python
# 제조사별 더 보기 버튼 클릭2
# Implicitly wait - indirect 접근 

#time.sleep(2)
#browser.find_element_by_xpath('//*[@id="dlMaker_simple"]/dd/div[2]/button[1]').click()
```

```python
# 원하는 모델카테고리 클릭 
# 애플 xpath   =    //*[@id="selectMaker_simple_priceCompare_A"]/li[16]/label


WebDriverWait(browser,2).until(EC.presence_of_element_located((By.XPATH,'//*[@id="selectMaker_simple_priceCompare_A"]/li[16]/label'))).click()
```

```python
# 2차 페이지 내용확인 
print('After Page Contents : {}'.format(browser.page_source))
```

```python
# 2초간 대기 
time.sleep(2)

# 현재 페이지 
cur_page = 1 

# 크롤링 페이지 수 
target_crwal_num = 5
```

```python


while cur_page <= target_crwal_num:
    # bs4 초기화 
    soup = BeautifulSoup(browser.page_source,'html.parser')

    # 소스 코드 정리
    #print(soup.prettify())    
    
    # 메인 상품 리스트 선택
    pro_list = soup.select('div.main_prodlist.main_prodlist_list > ul > li')

    # 상품 리스트 확인 
    #print(pro_list)

    # 페이지 번호 출력 
    print('****** Current Page num : {}'.format(cur_page),'******')
    print()
    
    
    # 필요 정보 추출 
    for v in pro_list:
        # 임시출력
        #print(v)

        if not v.find('div',class_='ad_header'):
            #상품명 출력
            #print(v.select('p.prod_name > a')[0])
            print(v.select('p.prod_name > a')[0].text.strip())

            # 이미지 출력
#             if not v.select('a.thumb_link > img')[0]['src']:             
#                 print(v.select('a.thumb_link > img')[0]['data-original'])
#             else:            
#                 print(v.select('a.thumb_link > img')[0]['src'])

            # 가격 출력
            print(v.select('p.price_sect > a')[0].text.strip())

        print()
    print()
    
    # 페이지 별 스크린샷 저장 
    
    # 페이지 증가 
    cur_page += 1
    
    if cur_page > target_crwal_num:
        print('종료')
        break
    
    # 페이지 이동 클릭 
    WebDriverWait(browser,3).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.number_wrap > a:nth-child({})'.format(cur_page)))).click()
    
    # 3초간 대기 
    time.sleep(3)
    
    
    
    # Beautifulsoup 인스턴스 삭제 
    del soup 
    
    
    
```

```python

# 브라우저 종료
browser.close()
```
