```python
# Httpbin 사이트를 이용한 Json 실습
#  - 수신 데이터 처리 
#  - 수신 데이터 - > Json 변환 출력 
#  - Response 다양한 정보 출력 

# Session04-2 
# Requests 
# Requests 사용 스크랩핑(2) - Json
import json
import requests
```

```python
s = requests.Session()

# 10 개 Json 데이터 요청 
r = s.get('https://httpbin.org/stream/10',stream = True )

# 수신 확인
print(r.text)

# Encoding 확인 
print('Before Encoding : {}'.format(r.encoding))


if r.encoding == None:
  r.encoding = 'UTF-8'

print('Afeter Encoding : {}'.format(r.encoding))


```

```python

for line in r.iter_lines(decode_unicode=True):
  # 라인 출력 후 타입 확인 
  print(line)
  print(type(line))


  # Json(DIct) 변환 후 타입 확인 
  b = json.loads(line) # str -> dict 변환 
  print(b)
  print(type(b))
  print("--------------")


  for k,v in b.items():
    print("Key : {}, Value : {}".format(k,v))

  print()
  print()

s.close()
```

```python
# 예제 2 'https://jsonplaceholder.typicode.com/todos/1' 


r = s.get('https://jsonplaceholder.typicode.com/todos/1')
# 헤더 정보
print(r.headers)

# 본문정보 
print(r.text)

# json 변환
print(r.json())

# key 반환 
print(r.json().keys())

# values 반환 
print(r.json().values())

# 인코딩 반환 
print(r.encoding)

# 바이너리 정보 
print(r.content)

s.close()

```

```python
# section 04-3 
# Requests 
# requests 사용 스크랩핑(3) - Rest API 

# Rest API : GET , POST , DELETE , PUT : UPDATE , REPLACE(FETECH : UPDATE ,MODIFY) - gpd put 
# 중요 : URL을 활용해서 resource 의 상태정보를 주고 받기 위해 

import requests
```

```python
# 세션 활성화 
s = requests.Session() 

# 예제 1  - get 사용
r = s.get('https://api.github.com/events')

# 수신상태 체크 
r.raise_for_status()

# 출력 
print(r.text)
```

```python
# 예제 3 - 쿠키 설정 ( 디테일하게 원하는 정보를 가져올때 )
jar = requests.cookies.RequestsCookieJar() 

# 쿠키 삽입 
jar.set('name','niceme', domain="httpbin.org",path='/cookies')

# 요청 
r = s.get('http://httpbin.org/cookies',cookies=jar)

# 출력 
print(r.text)
```

```python
# 예제3 - 사이트가 느릴때 5초여유 
r = s.get('http://github.com',timeout = 5 )

# 출력 
print(r.text)
```

```python
# 예제 4  - post 사용
r = s.post('http://httpbin.org/post',data = {'id':'test77','pw':'11'},cookies = jar )

print(r.text)
print(r.headers)
```

```python
# 예제 5 - post (요청) 2 
payload1 = {'id':'test77','pw':'11'}
payload2 = (('id','test778888'),('pw','18881'))

r = s.post('http://httpbin.org/post',data=payload1)
print( r.text)


r = s.post('http://httpbin.org/post',data=payload2)
print( r.text)
```

```python
# 예제 6 ( PUT ) 사용 

r = s.put('http://httpbin.org/put', data = payload1)
print(r.text)
```

```python
# 예제 7 (Delet) 

r = s.delete('http://httpbin.org/delete',data={'id':1})
print(r.text)
print()
print()


r = s.delete('https://jsonplaceholder.typicode.com/posts/1')
print(r.text)
print(r.ok)
print(r.headers)
s.close()


```
